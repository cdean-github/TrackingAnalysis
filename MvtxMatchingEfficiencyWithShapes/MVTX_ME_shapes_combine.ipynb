```python
import os
import pandas as pd
from IPython.display import HTML
```

```python
input_path = "/eos/user/j/jkvapil/SWAN_projects/sPHENIX"
file_list = ["df_shapes_pp_53534_p0001.pkl", "df_shapes_AuAu_54966_p0001.pkl"]
#real all dataframes
dfs = [pd.read_pickle(os.path.join(input_path, f)) for f in file_list]

# Define shared columns
merge_columns = ["shape", "nhits", "x", "y", "key", "layer"]

# Start with the first DataFrame
df_merged = dfs[0]

# Merge each subsequent DataFrame
for df in dfs[1:]:
    df_merged = pd.merge(
        df_merged,
        df,
        on=merge_columns,
        how='outer'  # Full outer join
    )
df_merged = df_merged.fillna("")
```

```python
html_table = df_merged.to_html(escape=False)
display(HTML(html_table))
```

```python
# Generate the HTML table
html_table = df_merged.to_html(escape=False, index=False)
prob = '0.' + file_list[0].split('_p')[-1].replace('.pkl', '')[1:]

# Add custom header
header_html = f"""
<h2 style="font-family:sans-serif;">Probability cutoff {prob}</h2>
"""

# Combine header + table
full_html = header_html + html_table

# Write to file
with open(f"{input_path}/PID_shapes_p{file_list[0].split('_p')[-1].replace('.pkl', '')}.html", "w") as f:
    f.write(full_html)
```
